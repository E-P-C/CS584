```python
import nltk
import pandas as pd
from nltk.sentiment import SentimentIntensityAnalyzer
# nltk.download('vader_lexicon') # Load the NLTK VADER model
```

```python
# Initialize VADER sentiment analyzer
sid = SentimentIntensityAnalyzer()
# Sample text for sentiment analysis
text = "ReaLLY Don't hate this product!"
# Perform sentiment analysis
sentiment_score = sid.polarity_scores(text)
# Print sentiment score
print((sentiment_score['compound'])) # -1~-0.05 neg; 0.05~1.0 positive
```

## Label jsonl files

```python
import json
```

```python
def label(text):
    sia = SentimentIntensityAnalyzer() # Only good on short text
    segments = ' '.join(text).split('\n')  # Split based on sentences
    scores = [sia.polarity_scores(segment)['compound'] for segment in segments if segment]
    
    if scores:
        avg_score = sum(scores) / len(scores)
        if avg_score > 0.05:
            return 'positive'
        elif avg_score < -0.05:
            return 'negative'
        else:
            return 'neutral'
    return 'neutral'  # Default if no segments were analyzed
```

```python
file_path = 'Data for candidate course project/news-snapchat-noagegroup-mental-2024.jsonl'
data = pd.read_json(file_path, lines=True)
print(data.keys())
```

```python
data['sentiment'] = data['body'].apply(label)

# Save to a new JSONL file
data[['title', 'body', 'provider', 'publication_date', 'sentiment']].to_json('labeled_snapchat.jsonl', orient='records', lines=True)

# Display the updated DataFrame with sentiments
print(data[['title', 'sentiment']][::29])
```
